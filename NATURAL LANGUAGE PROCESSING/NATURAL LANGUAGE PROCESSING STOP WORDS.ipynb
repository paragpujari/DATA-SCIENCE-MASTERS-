# Stop Words

These are the words that have no meaning and no significance to the text.

```python
# display all the arabic stop words
from nltk.corpus import stopwords

arabic_stopwords = set(stopwords.words('arabic'))

print(arabic_stopwords)
```

```python
# display all the english stop words
from nltk.corpus import stopwords

english_stopwords = set(stopwords.words("english"))

print(english_stopwords)
```

```python
# display all the spanish stop words

from nltk.corpus import stopwords

spanish_stopwords = set(stopwords.words('spanish'))

print(spanish_stopwords)
```

```python
# Remove all the stop words from the context

from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# import the raw file in the gutenberg corpus

sample = gutenberg.raw('carroll-alice.txt')
print(sample)
```

```python
# perform the word tokenization on the text

words = word_tokenize(sample)

print(words)
```

```python
filtered_words = []

for i in words:
    if(i not in english_stopwords):
        if(len(i) > 3):
            filtered_words.append(i)
            
print(filtered_words)
```

```python
# Get the frequency distribution of non stop words

from nltk.probability import FreqDist

freq_words = FreqDist(filtered_words)

print(freq_words)
```

```python
# Get the most common 20 words from the text
ans = freq_words.most_common(20)

print(ans)
```

```python
# plot the most common words from the text
freq_words.plot(20)
```
