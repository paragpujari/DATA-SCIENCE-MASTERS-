# Stemming

It refers to the process of reducing the words to their root words. The root words may be correct or may not be correct.

```python
# import the library for stemming

from nltk.stem import PorterStemmer

ps = PorterStemmer()

```

```python
# list of words for stemming
words = ["like", "likely", "liking", "unlike","liked"]
```

```python
# stem all the words to the root words
for i in words:
    print(i, ps.stem(i))
```

# Gutenberg Corpus

```python
from nltk.corpus import gutenberg
```

```python
# import the raw file in the gutenberg corpus
sample = gutenberg.raw('carroll-alice.txt')
print(sample)
```

```python
# perform the word tokenization on the above text
from nltk.tokenize import word_tokenize

words = word_tokenize(sample)
```

```python
print(words)
```

```python
# display all the stop words
from nltk.corpus import stopwords 

eng_stopwords = set(stopwords.words("english"))

print(eng_stopwords)
```

```python
# Remove all the stop words from the text
filtered_words = []

for i in words:
    if(i not in eng_stopwords):
        if(len(i) > 3):
            filtered_words.append(i)
            
print(filtered_words)
```

```python
# 1. Porter Stemmer

from nltk.stem import PorterStemmer

ps = PorterStemmer()
```

```python
# stem all the non stop words obtained from the text to their root words
for i in filtered_words:
    print(i, ps.stem(i))
```

```python
# 2. Snow ball Stemmer

from nltk.stem import SnowballStemmer
```

```python
snowstemmer = SnowballStemmer('english', ignore_stopwords = False)
```

```python
for i in words:
    print(i,snowstemmer.stem(i))
```
