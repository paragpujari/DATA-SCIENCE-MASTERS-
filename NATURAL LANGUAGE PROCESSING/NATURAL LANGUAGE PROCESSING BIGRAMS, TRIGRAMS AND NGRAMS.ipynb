# BIGRAMS

Set of two consecutive words taken at a time is called Bigrams.

```python
from nltk.corpus import webtext, stopwords

from nltk import bigrams
```

```python
# preprocess the data

words = [w.lower() for w in webtext.words('pirates.txt')]
print(words)
```

```python
# extract all the english stop words

english_stopwords = set(stopwords.words("english"))

print(english_stopwords)
```

```python
# filter out all the english stop words from the text

filtered_words = []

for i in words:
    if(i not in english_stopwords):
        if(len(i) > 3):
            filtered_words.append(i)
            
print(filtered_words)
```

```python
# Perform the Bigram operation

A = bigrams(filtered_words)
```

```python
print(A)
```

```python
# To find out the probability distribution of the filtered words
from nltk.probability import FreqDist


fre = FreqDist(A)

print(fre)
```

```python
# To find out the most common 20 words
most_common_words = fre.most_common(20)

print(most_common_words)
```

```python
# To plot the most common 20 words
fre.plot(20)
```

# TRIGRAMS

A set of three consecutive words is known as trigrams.

```python
from nltk import trigrams
```

```python
words = [w.lower() for w in webtext.words('grail.txt')]

print(words)
```

```python
# extract all the english stopwords

english_stopwords = set(stopwords.words("english"))

print(english_stopwords)
```

```python
# filter out all the english stop words from the text

filtered_words = []

for i in words:
    if(i not in english_stopwords):
        if(len(i) > 3):
            filtered_words.append(i)
            
print(filtered_words)
```

```python
# perform the trigram operation

B = trigrams(filtered_words)
```

```python
print(B)
```

```python
# To get the frequency distrinution of the data

from nltk.probability import FreqDist

ans = FreqDist(B)
```

```python
# To get the most common 20 words
ans.most_common(20)
```

```python
# To plot the most common 20 words

ans.plot(20)
```

# N GRAMS

Set of n continuous words in the text is called N-Grams

```python
from nltk.corpus import gutenberg, stopwords
```

```python
from nltk import ngrams
```

```python
text_words = [w.lower() for w in gutenberg.words('shakespeare-hamlet.txt')]
print(text_words)
```

```python
# extract all the english stop words
english_stopwords = set(stopwords.words("english"))
print(english_stopwords)
```

```python
# remove all the english stop words from the text
filtered_words = []

for i in text_words:
    if(i in english_stopwords):
        if(len(i) > 3):
            filtered_words.append(i)
            
print(filtered_words)
```

```python
# Apply N-Grams on the data
C = ngrams(filtered_words, n = 4)
```

```python
from nltk.probability import FreqDist

ans = FreqDist(C)
```

```python
# get the most common 30 words
ans.most_common(30)
```

```python
# plot the most common 30 words

ans.plot(30)
```
