```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Boston House Price Dataset
dataset = pd.read_csv('/content/BostonHousing.csv')
```

```python
print(dataset)
```

```python
# display the first five rows of the dataset
print(dataset.head())
```

```python
# display the last five rows of the dataset
print(dataset.tail())
```

```python
# depict the shape of the dataset
print(dataset.shape)
```

```python
# depict the information about all the columns in the dataset
print(dataset.info())
```

```python
# depict the statistical summary about the dataset
print(dataset.describe())
```

```python
# Divide the dataset into independent and dependent features
data_x = dataset.drop(columns = 'medv')
print(data_x)


data_y = dataset['medv']
print(data_y)
```

```python
# Divide the dataset into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data_x,data_y,test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
Y_train
```

** APPLY DECISION TREE REGRESSOR**

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor()

# train the model
regressor.fit(X_train, Y_train)
```

```python
# predict the model
pred_y = regressor.predict(X_test)
```

```python
print(pred_y)
```

```python
from sklearn.metrics import r2_score

score = r2_score(pred_y, Y_test)

print("Accuracy of the model is:", (score * 100.0))
```

**# APPLY GRID SEARCH CV ALGORITHM**

```python
# Define the parameters

parameter = {
    'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV

regressor_cv = GridSearchCV(regressor, param_grid = parameter, cv = 4, scoring = 'neg_mean_squared_error')
```

```python
# train the model
regressor_cv.fit(X_train, Y_train)
```

```python
# again train the model
regressor_cv.fit(X_train, Y_train)
```

```python
regressor_cv.best_params_
```

```python
# predict the model
pred_new_y = regressor_cv.predict(X_test)
print(pred_new_y)
```

```python
score = r2_score(pred_new_y, Y_test)
print("Accuracy of the model is", (score * 100.0))
```
