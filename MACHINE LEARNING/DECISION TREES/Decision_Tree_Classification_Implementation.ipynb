```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
print(df.head(3))
```

```python
# Divide the dataset into independent and dependent features
X = df.iloc[:,:-1]
X
```

```python
W = df.iloc[:,4]
```

```python
print(W)
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, W_train, W_test = train_test_split(X,W, test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
W_train
```

```python
W_test
```

**CONSTRUCT THE DECISION TREE MODEL**

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion = 'entropy')

# train the model
classifier.fit(X_train, W_train)
```

```python
from sklearn import tree
plt.figure(figsize = (12,10))
tree.plot_tree(classifier,filled = True)
```

```python
# Post Prunning
classifier = DecisionTreeClassifier(criterion = 'entropy', max_depth= 2)
```

```python
classifier.fit(X_train, W_train)
```

```python
from sklearn import tree
plt.figure(figsize = (12,10))
tree.plot_tree(classifier,filled = True)
```

```python
# predict the model
pred_w = classifier.predict(X_test)
```

```python
print(pred_w)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(W_test, pred_w)
print("Accuracy of the model is:", (score * 100.0))
```

```python
print(classification_report(W_test, pred_w))
```

**PREPRUNNING**

```python
# parameter
parameter = {
    'criterion' : ['gini','entropy','log_loss'],
    'splitter'  : ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# MODEL Creation
classifier = DecisionTreeClassifier(max_depth = 2)
```

```python
# Apply the hyperparameters on the model
cvgrid = GridSearchCV(classifier, param_grid = parameter, cv = 5, scoring = 'accuracy')
```

```python
# train the model
cvgrid.fit(X_train, W_train)
```

```python
# Find the best parameters
cvgrid.best_params_
```

```python
W_test
```

```python
# predict the model
predict_w = cvgrid.predict(X_test)
```

```python
print(predict_w)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(W_test, predict_w)
```

```python
print('Accuracy of the model is:', (score * 100.0))
```

```python
print(classification_report(predict_w,W_test))
```
