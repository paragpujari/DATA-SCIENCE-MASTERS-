```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# making the datasets
from sklearn.datasets import make_classification

X,y = make_classification(
    n_samples = 1000,
    n_features = 3,
    n_redundant = 1,
    n_classes = 2,
    random_state = 999
)
```

```python
X
```

```python
y
```

```python
# Divide the datasets into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

**APPLYING K-NEAREST CLASSIFIER ALGORITHM**

```python
from sklearn.neighbors import KNeighborsClassifier

kcn = KNeighborsClassifier(n_neighbors = 5, algorithm ='auto',p = 2)

kcn.fit(X_train, Y_train)
```

```python
# predict the model
pred_y = kcn.predict(X_test)
```

```python
print(pred_y)
```

**ACCURACY OF THE MODEL**

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
print(accuracy_score(pred_y, Y_test))
```

```python
print(confusion_matrix(pred_y, Y_test))
```

```python
print(classification_report(pred_y, Y_test))
```
