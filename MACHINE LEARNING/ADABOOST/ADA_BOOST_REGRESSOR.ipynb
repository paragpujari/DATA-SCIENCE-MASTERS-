```python
# import all the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/BostonHousing.csv')
```

```python
print(dataset)
```

```python
# display the first five rows of the dataset
print(dataset.head())
```

```python
# display the last five rows of the dataset
print(dataset.tail())
```

```python
# depict all the information about all the columns in the dataset
print(dataset.info())
```

```python
# divide the dataset into independent and dependent variables
data_x = dataset.drop(columns = 'medv')
print(data_x)

data_y = dataset['medv']
print(data_y)
```

```python
# Divide the dataset into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data_x,data_y,test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
# perform the Standard Scaling of the model
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

# train the model
X_train = sc.fit_transform(X_train)

X_test  = sc.transform(X_test)


print(X_train)

print(X_test)
```

**APPLY ADA BOOST REGRESSOR ALGORITHM**

```python
from sklearn.ensemble import AdaBoostRegressor

adareg = AdaBoostRegressor(loss='linear')

# train the model
adareg.fit(X_train, Y_train)
```

```python
# predict the model
pred_y = adareg.predict(X_test)
print(pred_y)
```

**ACCURACY OF THE MODEL**

```python
from sklearn.metrics import r2_score

ac = r2_score(pred_y, Y_test)

print("Accuracy of the model is:", (ac * 100.0))
```
