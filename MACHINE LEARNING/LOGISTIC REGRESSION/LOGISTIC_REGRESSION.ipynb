```python
# import all the neessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# loading the dataset
data = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
# display the entire dataset
print(data)
```

```python
# display the first five rows of the dataset
print(data.head())
```

```python
# display the last five rows of the dataset
print(data.tail())
```

```python
# display the shape of the dataset
print(data.shape)
```

```python
# display the detailed information about all the columns in the dataset
print(data.info())
```

```python
# display the statistical summary about all the infomrations in the dataset
print(data.describe())
```

```python
# divide the datset into independent and dependent variables

# independent variables
data_x = data.iloc[:,2:4].values
print(data_x)

data_y = data.iloc[:,4].values
print(data_y)
```

```python
# split the independent and dependent variables into training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(data_x, data_y, test_size = 0.25, random_state = 0)
```

```python
# perform the standard scaling process to bring all the inputs in one range
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

# transform the input data
X_train = sc.fit_transform(X_train)

X_test = sc.transform(X_test)
```

```python
X_train
```

**APPLY LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()

# train the model
log.fit(X_train, Y_train)
```

```python
# predict the model
Y_pred = log.predict(X_test)
```

```python
print(Y_pred)
```

```python
# actual output
print(Y_test)
```

```python
# finding out the cost function

from sklearn import metrics

print('Mean Absolute Error is:', metrics.mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error is:', metrics.mean_squared_error(Y_test, Y_pred))
print('Root Mean Squared Error is:', np.mean(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
# finding out the accuracy of the model
from sklearn.metrics import r2_score

score = r2_score(Y_pred, Y_test)

print('Accuracy of the model is:', (score * 100.0))
```

```python
# Plotting of Graph for Logistic Regression for training dataset
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:,0].min()-1, stop = X_set[:,0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:,1].min()-1, stop = X_set[:,1].max() + 1, step = 0.01))
plt.contourf(X1, X2, log.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.5, cmap = ListedColormap('red', 'green'))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0], X_set[y_set == j,1],
               c = ListedColormap(('red','green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
# Plotting of Graph for Logistic Regression for testing dataset
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:,0].min()-1, stop = X_set[:,0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:,1].min()-1, stop = X_set[:,1].max() + 1, step = 0.01))
plt.contourf(X1, X2, log.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
            alpha = 0.5, cmap = ListedColormap('red', 'green'))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j,0], X_set[y_set == j,1],
               c = ListedColormap(('red','green'))(i), label = j)
plt.title('Logistic Regression (Testing set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
