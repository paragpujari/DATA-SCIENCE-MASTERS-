```python
# import all the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the dataset
dataset = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
# display the first five rows of the dataset
print(dataset.head())
```

```python
print(dataset.info)
```

```python
print(dataset.describe())
```

```python
# divide the dataset into independent and dependent features

dataset_x = dataset.iloc[:,3:13]

dataset_y = dataset.iloc[:,13]
```

```python
print(dataset_x)
```

```python
print(dataset_y)
```

**FEATURE ENGINEERING**

```python
geography = pd.get_dummies(dataset['Geography'], drop_first = True)


gender    = pd.get_dummies(dataset['Gender'], drop_first = True)
```

```python
geography['Germany'] = geography['Germany'].astype('int')

geography['Spain'] = geography['Spain'].astype('int')

gender['Male'] = gender['Male'].astype('int')
```

```python
# concatenate the varibales with the dataframe

# first drop the columns
dataset_x = dataset_x.drop(['Geography','Gender'], axis = 1)
```

```python
dataset_x.head()
```

```python
dataset_x = pd.concat([dataset_x,geography, gender], axis = 1 )
```

```python
print(dataset_x.head())
```

```python
# spliiting the dataste into training and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(dataset_x, dataset_y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
X_test
```

**FEATURE SCALING**

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()


# train the model
X_train = sc.fit_transform(X_train)

X_test  = sc.transform(X_test) # transform is only used in test data to avoid data leakage
```

```python
X_train
```

```python
X_test
```

```python
X_test.shape
```

**ARTIFICIAL NEURAL NETWORKS**

```python
# import the libraries for Artificial Neural Networks

import keras
from   keras.models import Sequential
from   keras.layers import Dense
```

```python
from   keras.layers import Dropout
```

```python
# initialize the ann model
ann = Sequential()

# Adding the first input layer
ann.add(Dense(11,activation = 'relu'))


# Adding the first hidden layer
ann.add(Dense(7, activation = 'relu'))

# Adding the second hidden layer
ann.add(Dense(6, activation = 'relu'))

# Adding the output layer
ann.add(Dense(1, activation = 'sigmoid'))
```

```python
# compile the ANN Model
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
import tensorflow as tf

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
# Train the ANN Model
ann.fit(X_train, Y_train, validation_split = 0.33, batch_size = 10, epochs = 1000, callbacks = early_stopping)
```

```python
# prediction for the test data
pred_y = ann.predict(X_test)
```

```python
print(pred_y)
```

```python
pred_y = (pred_y >= 0.5)
```

```python
print(pred_y)
```

```python
# make the confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, pred_y)
```

```python
print(cm)
```

```python
# Accuracy of the model
from sklearn.metrics import accuracy_score

ac = accuracy_score(pred_y, Y_test)

print(ac)
```
