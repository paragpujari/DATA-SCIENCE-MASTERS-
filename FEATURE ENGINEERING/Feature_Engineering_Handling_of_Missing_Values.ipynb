**IMPUTATION MISSING TECHNIQUES**

**Mean Value Imputation**

It refers to the process of replacing all the null values with the mean of the particular column.

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.isnull().sum()
```

#** Handling the missing values of age column**

```python
# fill all the null values in the Age column with the mean

df['age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age_mean', 'age']]
```

**2. Median Imputation Technique**

It is used when we have outliers in the dataset.

**  This method works well when we have normally distributed data.**

```python
# fill all the null values with the median of that column
df['age_median'] = df['age'].fillna(df['age'].median())
```

```python
df.head()
```

```python
df[['age', 'age_mean', 'age_median']]
```

**3. Mode Imputation Technique**

It works well with categorical values

```python
df['embarked'].unique()
```

```python
# find the mode value of 'embarked'
df['embarked'].mode()
```

```python
modes_values = df['embarked'][0]
```

```python
print(modes_values)
```

```python
# replace all the null values in the embarked column with the mode value
df['embarked_modes'] = df['embarked'].fillna(df['embarked'].mode())
```

```python
df[['embarked_modes', 'embarked']]
```
