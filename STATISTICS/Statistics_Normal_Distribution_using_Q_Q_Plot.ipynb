```python
# importing all the necessary libraries
import scipy.stats as stat
import numpy as np
import pylab
import pandas as pd
```

```python
import seaborn as sns
```

```python
# loading the dataset
data = sns.load_dataset('iris')
```

```python
# print the first five records of the dataset
print(data.head())
```

```python
# print the last five records of the dataset
print(data.tail())
```

```python
import matplotlib.pyplot as plt
```

```python
# write a function to check whether every distribution follows the normal distribution or not
def plot_data(data, feature):
    # predicting the size of the figure
    plt.figure(figsize = (10,6))
    # getting the first subplot for the histogram
    plt.subplot(1,2,1)
    # plotting the histogram
    sns.histplot(data[feature] , kde = True)
    # To check whether the particular distribution follows the normal distribution or not
    stat.probplot(data[feature], dist = 'norm', plot = pylab)
    plt.show()
```

```python
plot_data(data, 'sepal_length')
```

```python
# OBSERVATION

##From the above data, we can see that most of the data points of 'sepal_length' are coming in straight line path, so the column 'sepal_length' follows the normal distribution.
```

```python
plot_data(data, 'sepal_width')
```

```python
# OBSERVATION

## From the above data, we can see that all the data points are falling in the straight line path, so the 'sepal_width' falls under the normal distribution.
```

```python
plot_data(data, 'petal_length')
```

```python
# OBSERVATION:

## From the above data, we can see that some of the data points do not come in the straight line path, so it does not follow the normal distribution.
```

```python
plot_data(data, 'petal_width')
```

```python
# OBSERVATION

## From the above data, we can see that the column 'petal_width' all the data points do not fall in the straight line path, so the 'petal_width' do not follow a normal distribution.
```
